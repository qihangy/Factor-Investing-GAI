```python
import pandas as pd
```

```python
df = pd.read_csv('winsor_factors_univariate.csv')
```

```python
columns = ['date', 'tic', 'b_m', 'e_p', 'cf_p', 'o', 'no', 'i_a', 
           'noa','pia', 'ig', 'nsi', 'ivg', 'ivc', 'oa', 'poa',
           'pta', 'gpa', 'oc_a', 'ol', 'R6', 'R11', 'sue',
           'roe', 'roa', 'nei']
df = df[columns]
```

```python
df1811 = df[df['date'] == '2018/11/30'].dropna().set_index('tic').drop('date', 1)
```

```python
from scipy import stats
```

```python
df_zscore = pd.DataFrame(stats.zscore(df1811, axis = 0), index = df1811.index, columns = df1811.columns)
```

```python
df_zscore.rank(ascending = False)
```
